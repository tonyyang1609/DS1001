```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import matplotlib
from sklearn.decomposition import PCA
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("movieReplicationSet - Copy.csv")
```

```python
for i in range(421,475):
    df.iloc[:,i].fillna(-1, inplace=True)
res = df.iloc[:,421:474]
```

```python
scaler = StandardScaler()
res=res.copy()
res=pd.DataFrame(scaler.fit_transform(res), columns=res.columns)
```

```python
print(res)
pca = PCA(n_components=0.5)
pca.fit(res)
print(pca.explained_variance_ratio_)
```

```python
PC_values = np.arange(pca.n_components_)+1
plt.plot(PC_values, pca.explained_variance_ratio_, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.show()
```

```python
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

loading_matrix = pd.DataFrame(loadings, index=res.columns)
loading_matrix
```

```python
res = pca.transform(res)
print(res)
```

```python
df_pca = pd.DataFrame(data=res)
print(df_pca)
```

```python
import seaborn as sns
sns.pairplot(df_pca)
```

```python
import copy

df = pd.read_csv("movieReplicationSet - Copy.csv")
data = df.iloc[:,:477]
label = []
feature = []
data = data.to_numpy()
for i in range(0,1097):
    buf = data[i,:477]
    for j in range(0,400):
        if(buf[j] >= 0):
            label.append(buf[j])
            temp = copy.deepcopy(data[i,:477]) - np.nanmean(buf)
            temp[j] = 0
            feature.append(np.nan_to_num(temp,nan=0))
```

```python
feature = np.array(feature)
label = np.array(label)
feature_train = feature[:int(0.8*(len(feature)))]
feature_test = feature[int(0.8*(len(feature))):]
label_train = label[:int(0.8*(len(feature)))]
label_test = label[int(0.8*(len(feature))):]
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
inputs = keras.Input(shape=(477,1))
x = layers.Dense(256, activation="relu")(inputs)
x = layers.Dense(128, activation="relu")(x)
x = layers.Dense(64, activation="relu")(x)
x = layers.Dense(32, activation="relu")(x)
x = layers.Dense(16, activation="relu")(x)

outputs = layers.Dense(1)(x)
```

```python
model = keras.Model(inputs=inputs, outputs=outputs, name="mnist_model")
```

```python
model.compile(
    loss="mse",
    optimizer="adam",
    metrics="mae",
)

history = model.fit(feature_train, label_train, batch_size=64, epochs=1, validation_split=0.2)
```

```python
a = np.mean(label)
sum = 0
for i in label:
    t = i - a
    if (t>0):
        sum = sum+t
    else:
        sum = sum - t
print(sum/len(label))
print(np.std(label))
```

```python
test_scores = model.evaluate(feature_test, label_test, verbose=2)
print("Test loss:", test_scores[0])
print("Test mae:", test_scores[1])
```
